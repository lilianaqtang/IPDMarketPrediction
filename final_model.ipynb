# Prepare the features

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.externals import joblib
```

```python
# Load the data set
df = pd.read_csv('C:/Users/ltang/Desktop/IPD prediction/final_dataset.csv', engine='python')
df.columns
```

```python
# Remove the fields from the data set that we don't want to include in our model
del df['ï»¿year']

# Replace categorical data with one-hot encoded data
features_df = pd.get_dummies(df, columns=['geo', 'structure_type'])

# Remove the number of building permits from the feature data
del features_df['building_permits_sum']

# Create the X and y arrays
X = features_df.as_matrix()
y = df['building_permits_sum'].as_matrix()
```

```python
# Split the data set in a training set (70%) and a test set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)

# Fit regression model
model = ensemble.GradientBoostingRegressor(
    n_estimators=1000,
    learning_rate=0.1,
    max_depth=6,
    min_samples_leaf=9,
    max_features = 0.1,
    loss = 'huber'
)

model.fit(X_train, y_train)
# Save the trained model to a file so we can use it in other programs
joblib.dump(model, 'trained_IPD_classifier_model.pkl')
```

```python
# Find the error rate on the training set
mse = mean_absolute_error(y_train, model.predict(X_train))
print("Training Set Mean Absolute Error: %.4f" % mse)

# Find the error rate on the test set
mse = mean_absolute_error(y_test, model.predict(X_test))
print("Test Set Mean Absolute Error: %.4f" % mse)
```
